```python
rows = 5

for i in range(rows):
    print(" " * (rows - i - 1) + "*" * (2 * i + 1))
```

```python
rows = 5
num = 1

for i in range(1, rows + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```
